```python
import pandas as pd

def spliting_record(record):
    city, rest_1 = record.split(' - ')
    city_code, rest_2 = rest_1.split(' (')
    Region, Region_code = rest_2.split(') [Region: ')
    Region_code = Region_code[:-1]  # Remove the closing bracket

    return city,city_code,Region,Region_code
```

```python
df[["city","city_code","Region","Region_code"]] = df['CITY_AND_REGION_CODES'].apply(lambda x: pd.Series(spliting_record(x)))
```
